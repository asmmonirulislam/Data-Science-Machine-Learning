```python
# Filtering ScopusAll according to Scopus's Title
```

```python
# converting the title into a list

scopus_title = scopus['Title'].to_list()
len(scopus_title)
```

```python
# Filtering ScopusAll

scopus_filtered = scopus_all[scopus_all['Title'].isin(scopus_title)].reset_index(drop=True)
len(scopus_filtered)
```

```python
# Droping the columns except Title, Year, DOI

scopus_filtered = scopus_filtered[['Title', 'Year', 'DOI']]
scopus_filtered.columns
```

```python
# Filtering the IEEEXploreAll according to the IEEEXploreSelected's title
```

```python
# Converting the titles into a list

ieee_title = ieee_selected['title'].to_list()
ieee_title
```

```python
len(ieee_title)
```

```python
# Filtering the IEEEXploreAll based on the created list

ieee_filtered = ieee_all[ieee_all['Title'].isin(ieee_title)].reset_index(drop=True)
ieee_filtered.count()   # got 9 papers
```

```python
# Droping all the columns except the Title, Year, and DOI

ieee_filtered = ieee_filtered[['Title', 'Year', 'DOI']]
ieee_filtered.info()
```

```python
# Converting the DataFrame into a CSV file

ieee_filtered.to_csv("IEEEXploreFiltered.csv", index=False)
```

```python
# Importing the newly created CSV file

try:
    scopus_filtered = pd.read_csv("ScopusFiltered.csv")
    ieee_filtered = pd.read_csv("IEEEXploreFiltered.csv")
except Exception as e:
    print(e)
else:
    print("Files Imported")
```

```python
# Adding a new column named: Database

scopus_filtered['Database']='Scopus'
science_direct['Database']='ScienceDirect'
springer['Database']='Springer'
ieee_filtered['Database']='IEEEXplore'
```

```python
# Calculating the number of papers

print(f"Scopus = {len(scopus_filtered)}")
print(f"ScienceDirect = {len(science_direct)}")
print(f"Springer = {len(springer)}")
print(f"IEEEXplore = {len(ieee_filtered)}")
print(f"Total = {len(scopus_filtered)+len(science_direct)+len(springer)+len(ieee_filtered)}")
```

```python
# Merging the files into one dataFrame

cols = ['Title', 'Year', 'DOI', 'Database']
dfs = [scopus_filtered[cols], science_direct[cols], springer[cols], ieee_filtered[cols]]

df = reduce(
    lambda x, y: pd.merge(
        x, y, on=cols, how='outer'
    ),
    dfs
)

len(df)
```

```python
# Removing the duplicates

df.drop_duplicates(subset=['Title'], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# Calculating the length of the merges file

print(f"Total Papers (After removing the duplicates): {len(df)}")
```

```python
# Checking the number of papers from 2022-2026

print(f"Papers from the last 5 years: {len(df[df['Year']>2021])}")
```

```python
# Removing the papers before 2022

df = df[df['Year']>=2022].reset_index(drop=True)
df
```

```python
# Sorting the papers based on 'Database', and 'Year'

df = df.sort_values(by=['Database', 'Year'], ascending=False).reset_index(drop=True)
```
