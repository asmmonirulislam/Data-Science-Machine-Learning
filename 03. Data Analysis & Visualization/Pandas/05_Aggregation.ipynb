```python
# Aggregate Function = Reduces a set of values into single summary value 
#                      Used to summarize and analyze data
#                      often used with the groupby() function
```

```python
import pandas as pd
try:
    df = pd.read_csv('../data.csv')
except Exception as e:
    print(e)
```

```python
# For Whole Data Frame
```

```python
df.mean(numeric_only=True) # returns the mean of all the numeric fields
```

```python
df.sum(numeric_only=True) # returns the sum of all the numeric fields
```

```python
df.std(numeric_only=True) # returns the standard deviation of all the numeric fields
```

```python
df.min(numeric_only=True) # returns the min value of all the numeric fields
```

```python
df.max(numeric_only=True) # returns the max value of all the numeric fields
```

```python
df.count() # returns the total count of all the columns
```

```python
# For Single Column
```

```python
df['Height'].mean()
```

```python
df['Height'].sum()
```

```python
df['Height'].std()
```

```python
df['Height'].min()
```

```python
df['Height'].max()
```

```python
df['Height'].count()
```

```python
# groupby() Function
```

```python
# What is the average height of the group by Type1?

group = df.groupby('Type1')
group['Height'].mean()
```

```python
# Total Height of the group by Type1

group['Height'].sum()
```

```python
# Min height of the group by Type1

group['Height'].min()
```

```python
# Max height of the group by Type1

group['Height'].max()
```

```python
# Count the pokemon by Type1

group['Height'].count()
```
