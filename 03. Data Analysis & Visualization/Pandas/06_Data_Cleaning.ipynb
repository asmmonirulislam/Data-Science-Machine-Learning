```python
# Data Cleaning = the process of fixing / removing incomplete, incorrect, or irrelevant data.
#                 ~75% of work done by pandas is Data Cleaning
```

```python
import pandas as pd
try:
    df = pd.read_csv('../data.csv')
except Exception as e:
    print(e)
```

```python
# 1. Drop irrelevant columns
```

```python
df = df.drop(columns=['No'])
df.head()
```

```python
# 2. Handle missing data
```

```python
pd.set_option('display.max_rows', None) # is used to show all the rows
df = df.dropna(subset=['Name', 'Type1', 'Type2', 'Height', 'Weight'])
df.count()
```

```python
# df # remove the # to view the entire table
```

```python
df = df.fillna({'Type2':'None'})
# It will fill the null value with "None"
# But this will not work now as the null values are already removed
```

```python
# 3. Fix inconsistent values
```

```python
df['Type1'] = df['Type1'].replace({'Grass':'GRASS', 'Fire':'FIRE'})
df['Type1'].unique().tolist()
```

```python
# 4. Standardize text
```

```python
df['Name'] = df['Name'].str.lower()
df['Name'].head()
```

```python
# 5. Fix Data Types
```

```python
# Change the data-type of Legendary to Boolean

df['Legendary'] = df['Legendary'].astype(bool)
df
```
