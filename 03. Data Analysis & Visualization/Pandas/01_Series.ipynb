```python
# Series = A Pandas 1-Dimensional labeled array that can hold any data type
# think of it like a single column in a spreadsheet (1-Dimensional)
```

```python
import pandas as pd
```

```python
# initialized the data list
data = [10, 20, 30]
```

```python
series = pd.Series(data) # Series() is a constructor
series # all data will be printed in a single column with the indexes at the left column
# output
# 0    10
# 1    20
# 2    30
# dtype: int64
```

```python
data = [10.5, 20.5, 30.5]
series = pd.Series(data)
series
# Output:
# 0    10.5
# 1    20.5
# 2    30.5
# dtype: float64 # data type changed
```

```python
data = ['A', 'B', 'C']
series = pd.Series(data)
series
# Output:
# 0    A
# 1    B
# 2    C
# dtype: object
```

```python
data = [True, False]
series = pd.Series(data)
series
# Output:
# 0     True
# 1    False
# dtype: bool
```

```python
data = [100, 200, 300, 400, 500]
```

```python
series = pd.Series(data, index=['number #1', 'number #2', 'number #3', 'number #4', 'number #5'])
series
```

```python
series = pd.Series(data, index=['a', 'b', 'c', 'd', 'e']) # index=[] is a list of indexes
series
```

```python
#loc[] = location by label

series.loc['a'] # returns the value at location 'a'
```

```python
# We can change the value of series by accessing with the loc[] property
series.loc['c']=350
series.loc['c'] # value of location 'c' changed from 300 to 350 
```

```python
# iloc[] = integer-location based indexing
series.iloc[2]
```

```python
# Boolean Filtering
series[series>200] # returns the value greater than 200 
```

```python
series[series<200]
```

```python
calories = {
    "day 1":1750,
    "day 2":2100,
    "day 3":1600
}
```

```python
series = pd.Series(calories)
series.loc['day 1']
```

```python
series.loc['day 3']+=500
series.loc['day 3'] # value of 'day 3' changed from 1600 to 2100
```
