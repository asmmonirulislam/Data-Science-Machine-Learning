```python
# DataFrame = A  tabular data structure with rows AND columns (2-Dimensional)
# Similar to an Excel Spreadsheet
```

```python
import pandas as pd
```

```python
data = {
    "Name" : ['student 1', 'student 2', 'student 3', 'student 4', 'student 5'],
    "Age" : [21, 22, 20, 19, 23],
    "CGPA" : [3.95, 3.56, 4.00, 3.19, 3.82]
}
data # data is a dictionary of lists
```

```python
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4', 's5'])
df
```

```python
df.loc['s2']
```

```python
df.iloc[1]
```

```python
df.iloc[2]
```

```python
# Add a new column
# df.insert(position, column_name, values)

df.insert(2, "Contact", ["01401411046", "01543543412", "01873123456", "01923123456", "01624123456"])
df
```

```python
df.info()
```

```python
# Add a new row

new_row = pd.DataFrame([
        {'Name':'student 6', 'Age':24, 'CGPA':3.45, 'Contact':'01302123456'},
        {'Name':'student 7', 'Age':21, 'CGPA':3.23, 'Contact':'01670123456'},
        {'Name':'student 7', 'Age':21, 'CGPA':3.23, 'Contact':'01670123456'},
    ], index=['s6', 's7', 's8'])
df = pd.concat([df, new_row])
df
```

```python
# Add a new row
# df.loc[index] = new_record

new_student = {
    "Name":"student 7",
    "Age":25,
    "CGPA":4.00,
    "Contact": "01234567891"
}

index = 's'+ str(len(df)+1)
df.loc[index] = new_student
df
```

```python
# Row = axis 0
# Column = axis 1
```

```python
# Remove a column
# df.drop("Column_name", axis=1, inplace=True)

df.drop("Age", axis=1, inplace=True)
df
```

```python
# Remove a row
# df.drop("row_index", axis=1, inplace=True)

df.drop('s1', axis=0, inplace=True)
df
```

```python
# Remove Duplicates

df.drop_duplicates(inplace=True)   # removes duplicates by comparing all the columns
df
```

```python
# Remove Duplicates

df.drop_duplicates(subset=["Name"])
```

```python
# Change a value

df.loc['s2', "CGPA"] = 2.50
df
```

```python
# Change the name of the Column
# df.rename(columns={"old_name": "new_name"}, inplace=True)

df.rename(columns={"Contact":"Phone"}, inplace=True)
df.info()
```

```python
# Sort DataFrame

df = df.sort_values(by="CGPA", ascending=False)
df
```

```python
df[df['CGPA']<3.50]
```

```python
# Convert the data into CSV File
df.to_csv('../students.csv', index=False)
```
