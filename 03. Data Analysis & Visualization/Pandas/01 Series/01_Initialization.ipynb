```python
# Series = A Pandas 1-Dimensional labeled array that can hold any data type
# think of it like a single column in a spreadsheet (1-Dimensional)
```

```python
import pandas as pd
```

```python
data = [10, 20, 30]
series = pd.Series(data)
series
```

```python
data = [10.5, 20.5, 30.5]
series = pd.Series(data)
series
```

```python
data = ['A', 'B', 'C']
series = pd.Series(data)
series
```

```python
data = [True, False]
series = pd.Series(data)
series
```
