```python
meanVal = df['Weight'].mean()
stdVal = df['Weight'].std()
lowerBound = meanVal - (3*stdVal)
upperBound = meanVal + (3*stdVal)
```

```python
print(f"Lower Bound: {lowerBound} and Upper Bound: {upperBound}")
```

```python
df
```

```python
df = df[(df['Weight'] >= lowerBound) & (df['Weight'] <= upperBound)]

# It removes all the rows whice has Weight less than lowerBound and greater than upperBound
```
