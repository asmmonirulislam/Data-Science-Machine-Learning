```python
# Filtering = Refers to the process of selecting elements
#             from an array that match a given condition
```

```python
import numpy as np
```

```python
ages = np.array([
    [21, 17, 19, 20, 16, 30, 18, 65],
    [39, 22, 15, 99, 18, 19, 20, 21]
])
```

```python
teenegers = ages[ages<18]
teenegers
```

```python
adults = ages[(ages>18) & (ages<=65)]
print(f"Adults are {adults}")
```

```python
even = ages[ages%2==0]
even
```

```python
# where(condition, array, fill) function [N.B.: The values do not match the condition will be replaced by 'fill']
```

```python
adults = np.where(ages>=18, ages, -1)
print(adults)
```
