```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, np.nan, 5, 6, np.nan, 7, np.nan, 8, 9, 10])
print(arr)
```

```python
# np.isnan() function
```

```python
print(np.isnan(arr)) # returns True for the NULL value; If there is a True in the array, it means there exists a NULL value
```

```python
arr = np.nan_to_num(arr, 0) # nan_to_num replaces the nan value with other numerical value
print(arr)
```

```python
# np.inf function
```

```python
arr = np.array([1, -2, 3, -np.inf, 5, 6, np.inf, np.inf])
print(arr)
```

```python
print(np.isinf(arr))
```

```python
arr = np.nan_to_num(arr, posinf=1000, neginf=-1000) # replaces positive inf to 1000 and negative inf to -1000
```
