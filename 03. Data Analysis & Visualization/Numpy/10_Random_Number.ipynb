```python
import numpy as np
```

```python
rng = np.random.default_rng()
```

```python
num = rng.integers(1,7)
print(num)
```

```python
num = rng.integers(low=1, high=7)
print(num)
```

```python
num = rng.integers(low=1, high=7, size=3) # 1D Array
print(num)
```

```python
num = rng.integers(low=1, high=7, size=(3,2)) # 2D Array
print(num)
```

```python
rng = np.random.default_rng(seed=1) # seed will produce the same output everytime 
num = rng.integers(low=1, high=7, size=(3,2))
print(num)
```

```python
# prints a random float number between 0 and 1
random = np.random.uniform()
random
```

```python
# prints the random floating number between lowe and upper bound
random = np.random.uniform(low=-10, high=1)
random
```

```python
random = np.random.randint(low=1, high=5, size=(3, 2, 4))
random
```
