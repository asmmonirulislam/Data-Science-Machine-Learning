```python
np.sqrt(arr) # square root of all the array element
```

```python
# round() = Rounds to nearest integer [4.6 → 5, 4.4 → 4]
# floor() = Always rounds down [4.9 → 4]
# ceil() = Always rounds up  [4.1 → 5]
```

```python
arr = np.array([1.34, 2.78, 3.50])
```

```python
np.round(arr) # 1.34 → 1, 2.78 → 3, and 3.50 → 4
```

```python
np.floor(arr) # 1.34 → 1, 2.78 → 2, and 3 → 4
```

```python
np.ceil(arr) # 1.34 → 2, 2.78 → 3, and 3.50 → 4
```
