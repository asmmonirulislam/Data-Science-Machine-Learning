```python
# Bar Chart = Compare categories of data by representing each category with a bar
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = np.array(['Fruits', 'Grains', 'Vegetables', 'Protein', 'Dairy', 'Sweet'])
values = np.array([20, 10, 35, 25, 15, 40])
```

```python
plt.bar(categories, values, color='red')
plt.title('Daily Consumption')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```
