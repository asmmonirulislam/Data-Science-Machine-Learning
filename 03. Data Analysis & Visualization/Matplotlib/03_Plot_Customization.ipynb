```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
style1 = {
    'marker':'.',                         # marker shapes
    'markersize':'10',                    # marker size
    'markerfacecolor':'#ed0c0c',        # marker face color
    'markeredgecolor':'#000000',        # marker edge color
    'linestyle':'dashdot',                # style of the line
    'linewidth':2,                        # width of line
    'color':'#f26bf2'                   # color of the line
}

style2 = {
    'marker':'.',                         # marker shapes
    'markersize':'10',                    # marker size
    'markerfacecolor':"#e2ed0c",        # marker face color
    'markeredgecolor':'#000000',        # marker edge color
    'linestyle':'dashdot',                # style of the line
    'linewidth':2,                        # width of line
    'color':"#ee0606"                   # color of the line
}

title_style = dict(
    fontsize = 20,
    family = 'Arial',
    fontweight = 'bold',
    color = '#2d4cfc'
)
```

```python
x = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026]) # X-axis
y1 = np.array([10, 20, 30, 50, 35, 20, 40]) # Y-axis
y2 = np.array([15, 10, 35, 45, 19, 14, 40]) # Y-axis
```

```python
plt.plot(x, y1, **style1) # **kwargs (unpacking operator)
plt.show()
```

```python
plt.plot(x, y1, **style1)
plt.plot(x, y2, **style2)
plt.show()
```

```python
plt.plot(x, y1, **style1)
plt.plot(x, y2, **style2)

plt.title('Students of the Class', title_style)
plt.ylabel('Number of Students', title_style)
plt.xlabel('Year (2020-2026)', title_style)

plt.tick_params(
    axis='both',
    colors='red'
)

plt.show()
```

```python
# grid() function
```

```python
gridStyle = dict(
    axis='both',
    linewidth=1,
    color='yellow',
    linestyle='dashed'
)
```

```python
plt.plot(x, y1, **style1)
plt.plot(x, y2, **style2)

plt.title('Students of the Class', title_style)
plt.ylabel('Number of Students', title_style)
plt.xlabel('Year (2020-2026)', title_style)

plt.tick_params(
    axis='both',
    colors='red'
)

plt.grid(**gridStyle)

plt.show()
```
