```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = np.array(["Freshmen", "Sophomores", "juniors", "Seniors"])
values = np.array([300, 250, 275, 225])


colors = ['red', 'green', 'blue', 'yellow']
```

```python
plt.pie(values, 
        labels=categories, 
        autopct="%1.1f%%",  # %1.1f means a floating point number with 1 decimal value after the point
        colors=colors,
        explode=[0, 0, 0, 0.1] # explode is used to separate (pull out) one or more slices from the pie chart to emphasize them.
        
        ) 
plt.title('My College')
plt.show()
```
